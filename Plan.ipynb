# PROJECT PYTHON

## Objective

- Find the best spots for placing advertisements in the Prague public transportation system
- Do so for several characteristics
- The output: a plot of the best spots, a table showing the top resutls

## Defining "best spots"

1. High visibility - most passengers pass by this stop
2. Longest exposure
3. High repetition - frequent routes/trips
4. High accessibility location - touristy areas/ city centre

## Plan of action

### High visibility

- No passengers count available - use trips frequency as a proxy
- Count the number of unique trips
- Highest amount of unique trips per stop = highest visibility
- Further analysis for different types of transportation (metro, bus, tram), wheelchair and bike accessibility
- Datasets: stops_times.txt, stops.txt

### Longest exposure

- Two approaches: a trip lasts the longest time, or a trip has the biggest number of stops

1. __Time-wise approach:__
    - For each trip, find the departure time from the first stop and the arrival time at the last stop
    - Compute the difference
    - Dataset: stops.txt, trips.txt

2. __Number of stops approach:__
    - For each trip, count the number of stops
    - Dataset: stops.txt, trips.txt

- Further analysis for different types of transportation (metro, bus, tram), wheelchair and bike accessibility

### High repetition

- Count the number of trips per route
- Optionally, analyse for weekdays and weekends separately
- Further analysis for different types of transportation (metro, bus, tram), wheelchair and bike accessibility
- Dataset: trips.txt, routes.txt, calendar.txt

### High accessibility location

- Two approaches to finding stops in the best location

1. __Approach 1:__
    - Define a centre point of Prague (such as Václavské náměstí or Staroměstské náměstí)
    - Compute the distance of each stop from the centre point
    - Filter for distance smaller than some predefined value

2. __Approach 2:__
    - Compute a list of popular tourist places and/or a list of the biggest stadiums, train stations, the airport
    - Find their coordinates
    - Compute a "buffer" area around each place
    - Find the stops in the "buffer"

3. __Analysis:__
    - Rank by frequency - which high-tourist-area stops have the most trips
    - Find if some routes pass through more than one of these stops - highly used connections
    - Filter by accessibility - wheelchair-accessible stops, bike-accessible stops

- Further analysis for different types of transportation (metro, bus, tram), wheelchair and bike accessibility
- Dataset: stops.txt, stops_times.txt, routes.txt
